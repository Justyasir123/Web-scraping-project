```python
df['Salary']=df['Salary'].replace("Salary not disclosed",np.nan)
df.head(10)
```

```python
df['Salary'].value_counts()
```

```python
df.isnull().sum()
```

```python
df.dropna(inplace=True)
df.isnull().sum()
```

```python
df["Salary"].unique()
```

```python
# Split the Salary column into "Salary Min" and "Salary Max"
df["Salary"] = df['Salary'].str.split('-', expand=True)[1].str.strip()

df["Salary"]
```

```python
df["Salary"].value_counts()
```

```python
df["Salary"]=df["Salary"].str.replace("2000000 Yearly",str(2000000/12))
df["Salary"]=df["Salary"].str.replace("800000 Yearly",str(800000/12))
df["Salary"]=df["Salary"].str.replace("400000 Yearly",str(400000/12))
df["Salary"]=df["Salary"].str.replace("361000 Yearly",str(361000/12))
df["Salary"]=df["Salary"].str.replace("200000 Yearly",str(200000/12))

df["Salary"].unique()
```

```python
df["Salary"].value_counts()
```

```python
df["Salary"]=df["Salary"].str.split(" ").str[0]
```

```python
df["Salary"].unique()
```

```python
df["Salary"]=df["Salary"].str.split(".").str[0]
```

```python
df['Salary'] = pd.to_numeric(df['Salary'], errors='coerce').astype('Int64')
```
